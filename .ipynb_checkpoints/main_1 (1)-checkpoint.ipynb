```python
# Import des librairies n√©cessaires
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import fastcluster
from scipy.cluster.hierarchy import dendrogram, fcluster

# Cr√©ation d'une copie du dataframe sous le nom df_non_superviser
df_non_superviserT = df_model.copy()

# V√©rifiez si les colonnes existent avant de les supprimer
for column in ['Type de voie', 'Nature mutation']:
    if column in df_non_superviserT.columns:
        df_non_superviserT = df_non_superviserT.drop([column], axis=1)
        
# Filtrer pour obtenir uniquement les entr√©es de l'ann√©e 2022
df_non_superviserT = df_non_superviserT.loc[df_non_superviserT.index.year == 2022]

# Normalisation des donn√©es
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_non_superviserT)

# Cr√©ation de la matrice de linkage
Z = fastcluster.linkage_vector(df_scaled, method='ward')

plt.figure(figsize=(10, 7))
plt.title("Dendrogramme")
dendrogram(Z, truncate_mode='lastp', p=50)  # Affiche seulement les 50 derni√®res branches fusionn√©es
plt.show()

# D√©termination des clusters √† partir de la matrice de linkage
labels = fcluster(Z, t=5, criterion='maxclust')

# Ajouter les clusters au dataframe
df_non_superviserT['Cluster'] = labels 
```

```python
# Import des librairies n√©cessaires
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import fastcluster
from scipy.cluster.hierarchy import dendrogram, fcluster

# Cr√©ation d'une copie du dataframe sous le nom df_non_superviser
df_non_superviser = df_model.copy()

# V√©rifiez si les colonnes existent avant de les supprimer
for column in ['Type de voie', 'Nature mutation', 'Nombre de lots', 'Surface terrain']:
    if column in df_non_superviser.columns:
        df_non_superviser = df_non_superviser.drop([column], axis=1)
        
# Filtrer pour obtenir uniquement les entr√©es de l'ann√©e 2022
df_non_superviser = df_non_superviser.loc[df_non_superviser.index.year == 2022]
```

```python
# Normalisation des donn√©es
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_non_superviser)

# Cr√©ation de la matrice de linkage
Z = fastcluster.linkage_vector(df_scaled, method='ward')

plt.figure(figsize=(10, 7))
plt.title("Dendrogramme")
dendrogram(Z, truncate_mode='lastp', p=50)  # Affiche seulement les 50 derni√®res branches fusionn√©es
plt.show()

# D√©termination des clusters √† partir de la matrice de linkage
labels = fcluster(Z, t=5, criterion='maxclust')

# Ajouter les clusters au dataframe
df_non_superviser['Cluster'] = labels    
```

```python
# Calcul de l'IQR pour 'Valeur fonciere'
Q1 = df_non_superviser['Valeur fonciere'].quantile(0.25)
Q3 = df_non_superviser['Valeur fonciere'].quantile(0.75)
IQR = Q3 - Q1

# D√©finition des limites pour les valeurs aberrantes
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Filtrage des valeurs aberrantes
df_filtered = df_non_superviser[(df_non_superviser['Valeur fonciere'] >= lower_limit) & (df_non_superviser['Valeur fonciere'] <= upper_limit)]
```

```python
# Import des librairies n√©cessaires
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import fastcluster
from scipy.cluster.hierarchy import dendrogram, fcluster

# Cr√©ation d'une copie du dataframe sous le nom df_non_superviser
df_non_superviserALL = df_model.copy()

# V√©rifiez si les colonnes existent avant de les supprimer
for column in ['Type de voie', 'Nature mutation']:
    if column in df_non_superviserALL.columns:
        df_non_superviserALL = df_non_superviserALL.drop([column], axis=1)
        
# Normalisation des donn√©es
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_non_superviserALL)

# Cr√©ation de la matrice de linkage
Z = fastcluster.linkage_vector(df_scaled, method='ward')

# Visualisation du dendrogramme
plt.figure(figsize=(10, 7))
plt.title("Dendrogramme")
dendrogram(Z, truncate_mode='lastp', p=50)  # Affiche seulement les 50 derni√®res branches fusionn√©es
plt.show()

# D√©termination des clusters √† partir de la matrice de linkage
labels = fcluster(Z, t=5, criterion='maxclust')

# Ajouter les clusters au dataframe
df_non_superviserALL['Cluster'] = labels    
```

```python
def filter_aberation(df, column):
    """Filtre les valeurs aberrantes d'une colonne sp√©cifique dans un DataFrame"""
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    df_filtered = df[~((df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR)))]
    return df_filtered


def analyse_cluster(df, column):
    """Analyse une colonne sp√©cifique dans un DataFrame group√© par clusters"""
    # Calcule la moyenne pour chaque cluster
    mean_values = df.groupby('Cluster')[column].mean()
    print(f"Moyenne de '{column}' pour chaque cluster:")
    print(mean_values)

    # Calcule la m√©diane pour chaque cluster
    median_values = df.groupby('Cluster')[column].median()
    print(f"\nM√©diane de '{column}' pour chaque cluster:")
    print(median_values)

    # Calcule l'√©cart-type pour chaque cluster
    std_values = df.groupby('Cluster')[column].std()
    print(f"\n√âcart-type de '{column}' pour chaque cluster:")
    print(std_values)

    # Cr√©e un boxplot de la distribution par cluster
    plt.figure(figsize=(10, 6))
    sns.boxplot(x="Cluster", y=column, data=df)
    plt.title(f"Distribution de '{column}' pour chaque cluster")
    plt.show()
```

### 4.2.1) Modelisation

```python
def plot_predictions(y, y_prd, regressor):
    plt.scatter(y, y_pred)
    plt.title('title')
    plt.xlabel('Valeurs actuelles')
    plt.ylabel('Valeurs pr√©dites')
    plt.title(f'Predicted vs Actual Values ({regressor.__class__.__name__})')
    plt.axline((0,0), slope=1, color='k', linestyle='--', label='y=x', linewidth=0.8)
    max_xy = max(plt.xlim()[1], plt.ylim()[1])
    plt.xlim(0, max_xy)  
    plt.ylim(0, max_xy)  
    plt.gca().set_aspect('equal')
    plt.show()
```

```python
def print_score(R2, RMSE, regressor):
    display(Markdown(f'**R^2 score: {R2:.3f} for {regressor.__class__.__name__}**'))
    display(Markdown(f'**RMSE score: {RMSE:.3f} for {regressor.__class__.__name__}**'))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer, StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

# Les colonnes √† traiter
num_cols = ['Nombre de lots', 'Surface reelle bati', 'Nombre pieces principales', 'Surface terrain']
cat_cols = ['Type de voie', 'Nature mutation']

# Les cat√©gories pour le codage OneHot
cols_type_voie = ['RUE', 'AV', 'RTE', 'ALL', 'CHE', 'BD', 'PL', 'RES', 'CITE', 'IMP', 'VC']
cols_nature_mutation = ['Vente', 'Echange', 'Adjudication']

# Le pipeline pour les donn√©es num√©riques
num_transformer = ColumnTransformer(
    transformers=[
        ('passthrough', 'passthrough', ['Nombre de lots', 'Nombre pieces principales']),
        ('scaler', StandardScaler(), ['Surface reelle bati', 'Surface terrain']),
    ],
    remainder='drop'
)

# Le pipeline pour les donn√©es cat√©gorielles
cat_pipeline = Pipeline([
    ('onehot', OneHotEncoder(categories=[cols_type_voie, cols_nature_mutation], handle_unknown='ignore')),
])

# Combiner les deux pipelines
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_cols),
        ('cat', cat_pipeline, cat_cols)
    ])

# Liste des mod√®les √† tester
regressors = [
    LinearRegression(),
    RandomForestRegressor(),
    GradientBoostingRegressor(),
]

# Liste pour stocker les pr√©dictions de chaque mod√®le
predictions = []

# Boucle pour tester les mod√®les
for regressor in regressors:
    
    display(Markdown(f'# {regressor.__class__.__name__}'))
    
    # Cr√©er le pipeline final avec le pr√©processeur et le mod√®le de r√©gression
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('regressor', regressor)
    ])

    # Entra√Æner le mod√®le
    X = df_model.drop('Valeur fonciere', axis=1)
    y = df_model['Valeur fonciere']
    display(pipeline.fit(X, y))

    # Pr√©dictions
    y_pred = pipeline.predict(X)

    # Afficher l'importance des variables pour les mod√®les d'arbre
    if isinstance(regressor, (RandomForestRegressor, GradientBoostingRegressor)):
        feature_importances = regressor.feature_importances_
        feature_names = num_cols + cat_pipeline.named_steps['onehot'].get_feature_names_out(cat_cols).tolist()

        # Afficher l'importance des variables sous forme de graphique
        plt.figure(figsize=(8, 6))
        plt.barh(feature_names, feature_importances)
        plt.xlabel('Importance')
        plt.ylabel('Variable')
        plt.title(f'Feature Importance - {regressor}')
        plt.show()
        
        plot_predictions(y, y_pred, regressor.__class__.__name__)

    # Afficher les coefficients pour la r√©gression lin√©aire
    if isinstance(regressor, LinearRegression):
        cat_pipeline.fit(X[cat_cols])
        feature_names = num_cols + cat_pipeline.named_steps['onehot'].get_feature_names_out(cat_cols).tolist()
        coefficients = regressor.coef_

        # Afficher les coefficients sous forme de graphique
        plt.figure(figsize=(8, 6))
        plt.barh(feature_names, coefficients)
        plt.xlabel('Coefficient Value')
        plt.ylabel('Variable')
        plt.title(f'Coefficient Analysis - {regressor.__class__.__name__}')
        plt.show()
        
        plot_predictions(y, y_pred, regressor)

    # Calculer le score R^2
    R2 = pipeline.score(X, y)

    # Calculer le score RMSE
    RMSE = np.sqrt(mean_squared_error(y, y_pred))
    print_score(R2, RMSE, regressor)
```
