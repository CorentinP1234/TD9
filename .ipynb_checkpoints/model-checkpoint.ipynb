2. 'AV' : Avenue - une large rue, souvent bordée d'arbres.
3. 'RTE' : Route - une voie routière.
4. 'ALL' : Allée - une petite voie, souvent bordée d'arbres.
5. 'CHE' : Chemin - une petite voie, souvent en milieu rural.
6. 'BD' : Boulevard - une large avenue, souvent bordée d'arbres.
7. 'PL' : Place - une voie souvent entourée de bâtiments et où plusieurs rues se rencontrent.
8. 'RES' : Résidence - une zone résidentielle, souvent un ensemble de maisons ou d'appartements.
9. 'CITE' : Cité - une zone résidentielle, souvent un ensemble de logements sociaux.
10. 'IMP' : Impasse - une rue sans issue.
11. 'VC' : Voie Communale - une route gérée par une commune.
12. 'SQ' : Square - un petit parc ou un espace vert en milieu urbain.
13. 'RLE' : Ruelle - une petite rue étroite.
14. 'GR' : Grand Rue - désigne une rue principale d'une ville ou d'un village.
15. 'CRS' : Cours - une large voie, souvent une promenade ou une avenue.
